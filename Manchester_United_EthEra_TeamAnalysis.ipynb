```python
# XG(Expected goals) and XGA(Expected Goals Agains) Analysis
Eth_era['xG'] = pd.to_numeric(Eth_era['xG'], errors='coerce')
Eth_era['xGA'] = pd.to_numeric(Eth_era['xGA'], errors='coerce')

```

```python
total_xG = Eth_era['xG'].sum()
total_xGA = Eth_era['xGA'].sum()
avg_xG = Eth_era['xG'].mean()
avg_xGA = Eth_era['xGA'].mean()

print(f"Total xG: {total_xG}, Total xGA: {total_xGA}")
print(f"Average xG per match: {avg_xG}, Average xGA per match: {avg_xGA}")
```
